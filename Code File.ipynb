## Getting and pulling relevant data

1. "sale_listings_monthly" : is new unique listings by month, processed from zillow, by MSA
2. "sale_listing_weekly" : is new unique listings by week, processed from zillow, by MSA
3. "for_sale_inventory" : is total number of units on the market by week, processed from zillow, by MSA
4. "rent_monthly" : rental information by week, processed from apartments list


```python
sale_listings_monthly = pd.read_csv("https://files.zillowstatic.com/research/public_csvs/new_listings/Metro_new_listings_uc_sfrcondo_month.csv?t=1742519005")

sale_listing_weekly = pd.read_csv("https://files.zillowstatic.com/research/public_csvs/new_listings/Metro_new_listings_uc_sfrcondo_week.csv?t=1742519005")

for_sale_inventory = pd.read_csv("https://files.zillowstatic.com/research/public_csvs/invt_fs/Metro_invt_fs_uc_sfrcondo_week.csv?t=1742519005")

rent_monthly = pd.read_csv(os.path.join(dir, "Apartment_List_Rent_Growth_MoM_2025_02.csv"))
```

## Developing visualizations (box plots)

```python
cols = ['2025-01-18', '2025-01-25',
       '2025-02-01', '2025-02-08', '2025-02-15', '2025-02-22', '2025-03-01']

# Extract Washington DC values for the specified dates.
dc_vals = weekly_listing_pct.loc["Washington, DC", cols]

# For large cities, take the top 50 rows (ordered by population) excluding Washington DC.
df_large = weekly_listing_pct.drop("Washington, DC", errors="ignore").iloc[:50]

# For all cities, exclude Washington DC.
df_all = weekly_listing_pct.drop("Washington, DC", errors="ignore")
```

```python
def plot_box(df, title, ax, cols, dc_vals):
    data = [df[col].dropna() for col in cols]
    
    positions = np.arange(1, len(cols) + 1)
    bp = ax.boxplot(
        data,
        patch_artist=True,
        positions=positions,
        widths=0.6,
        showfliers=True
    )
    
    # Color boxes & medians
    for box in bp['boxes']:
        box.set(facecolor='lightblue', alpha=0.7, linewidth=1.5)
    for median in bp['medians']:
        median.set(color='orange', linewidth=2)

    # Set X ticks & labels to align with the box positions
    ax.set_xticks(positions)
    ax.set_xticklabels(
        [pd.to_datetime(dt).strftime('%b %d') for dt in cols],
        rotation=0,
        ha='center'
    )
    ax.set_xlim(0.5, len(cols) + 0.5)

    ax.set_title(title)
    ax.set_ylabel("Percentage Change")
    ax.grid(axis='y', linestyle='--', alpha=0.6)

    # Overlay DCâ€™s values
    for i, col in enumerate(cols):
        ax.scatter(
            positions[i],
            dc_vals[col],
            color='red',
            s=80,
            edgecolor='black',
            zorder=3,
            label="DC" if i == 0 else ""
        )

    # Create legend 
    handles, labels = ax.get_legend_handles_labels()
    if "DC" in labels:
        ax.legend(loc="upper left")

# 1) Top 50 Large Cities
fig1, ax1 = plt.subplots(figsize=(7,5))
plot_box(df_large, "New Inventory (%) by Week in Top 50 Cities", ax1, cols, dc_vals)
plt.tight_layout()
plt.ylim(-40,65)
plt.savefig(os.path.join(new_inv_dir, 'New Inventory_Box_Top 50 Cities.png'))
plt.show()

# 2) All Cities
fig2, ax2 = plt.subplots(figsize=(7,5))
plot_box(df_all, "All Cities", ax2, cols, dc_vals)
plt.tight_layout()
plt.ylim(-60,130)
plt.savefig(os.path.join(new_inv_dir,'New Inventory_Box_Top All Cities.png'))
plt.show()
```

## Heatmap

```python
cols = [
    '2025-01-18','2025-01-25','2025-02-01',
    '2025-02-08','2025-02-15','2025-02-22','2025-03-01'
]
labels = [pd.to_datetime(c).strftime('%b %d') for c in cols]

# Subset your data (rows 1:20, for instance)
df_heat = weekly_listing_pct[cols].iloc[1:15]

# Create heatmap
fig, ax = plt.subplots()
sns.heatmap(df_heat, 
            cmap="RdBu", 
            center=0, 
            linewidths=0.3, 
            xticklabels=labels, 
            ax=ax)
plt.xticks(rotation=90, ha='right')
plt.title("New Inventory Percentage Changes Heatmap")
plt.xlabel("Date")
plt.ylabel("City")

# Highlight Washington DC by drawing a rectangular outline
if 'Washington, DC' in df_heat.index:
    dc_row = df_heat.index.get_loc('Washington, DC')
    # x=0  (start at left edge), y=dc_row (DC row), width=len(cols), height=1
    rect = plt.Rectangle(
        (0, dc_row), len(cols), 1,
        fill=False, edgecolor='black', linewidth=1, linestyle = '--'
    )
    ax.add_patch(rect)

plt.tight_layout()
plt.savefig(os.path.join(new_inv_dir,'New Inventory_Heatmap_Top 15 Cities.png'))
plt.show()
```

## Line Graph for DC - Seperate by Years for 10 Weeks

```python
# 1) Extract the DC series, ensure datetime index
dc_series = weekly_listing_pct.loc["Washington, DC"].copy()
dc_series.index = pd.to_datetime(dc_series.index)
dc_series.sort_index(inplace=True)

# 2) Build a DataFrame so we can pull year/week easily
df = dc_series.to_frame(name='pct_change')
df['year'] = df.index.year
df['week'] = df.index.isocalendar().week

# Filter to the first 10 ISO weeks (i.e., week numbers 1..10)
df_1_10 = df[df['week'].between(1, 10)]

# 3) Create small multiples
unique_years = sorted(df_1_10['year'].unique())
n_years = len(unique_years)
n_cols = 4  
n_rows = math.ceil(n_years / n_cols)

fig, axes = plt.subplots(n_rows, n_cols, figsize=(4.5 * n_cols, 3 * n_rows), sharey=True)
axes = axes.flatten()  # Easier to iterate over a 1D list of axes

for ax, yr in zip(axes, unique_years):
    # Subset for this year
    sub = df_1_10[df_1_10['year'] == yr].copy()
    ax.plot(sub['week'], sub['pct_change'], marker='o')
    ax.set_title(f"{yr}")
    ax.set_xticks(range(1, 11))
    ax.set_xlabel("Week of Year")
    ax.set_ylabel("Pct Change")
    ax.grid(True, linestyle='--', alpha=0.5)

# Hide any unused subplots (if we have fewer years than n_rows*n_cols)
for j in range(len(unique_years), n_rows * n_cols):
    fig.delaxes(axes[j])

fig.suptitle("Washington DC: First 10 Weeks Inventory % Changes by Year", fontsize=14)
plt.tight_layout()
plt.show()
```

```python
def plot_box(df, title, ax, cols, dc_vals):
    # Collect column data into lists
    data = [df[col].dropna() for col in cols]
    
    # Box plot positions 1..N
    positions = np.arange(1, len(cols) + 1)
    bp = ax.boxplot(
        data,
        patch_artist=True,
        positions=positions,
        widths=0.6,
        showfliers=True
    )
    
    # Color boxes & medians
    for box in bp['boxes']:
        box.set(facecolor='lightblue', alpha=0.7, linewidth=1.5)
    for median in bp['medians']:
        median.set(color='orange', linewidth=2)

    # Set X ticks & labels to align with the box positions
    ax.set_xticks(positions)
    ax.set_xticklabels(
        [pd.to_datetime(dt).strftime('%b %d') for dt in cols],
        rotation=0,
        ha='center'
    )
    ax.set_xlim(0.5, len(cols) + 0.5)

    ax.set_title(title)
    ax.set_ylabel("Percentage Change")
    ax.grid(axis='y', linestyle='--', alpha=0.6)

    # Overlay DCâ€™s values
    for i, col in enumerate(cols):
        ax.scatter(
            positions[i],
            dc_vals[col],
            color='red',
            s=80,
            edgecolor='black',
            zorder=3,
            label="DC" if i == 0 else ""
        )

    # Create legend 
    handles, labels = ax.get_legend_handles_labels()
    if "DC" in labels:
        ax.legend(loc="upper left")

# 1) Top 50 Large Cities
fig1, ax1 = plt.subplots(figsize=(7,5))
plot_box(df_large, "On Market Inventory in Top 50 Large Cities", ax1, cols, dc_vals)
plt.tight_layout()
plt.savefig('Agg Inventory_Box_All Cities.png')
plt.show()

# 2) All Cities
fig2, ax2 = plt.subplots(figsize=(7,5))
plot_box(df_all, "On Market Inventory in All Cities", ax2, cols, dc_vals)
plt.tight_layout()
# plt.ylim(-60,130)
plt.savefig(os.path.join(tot_inv_dir,'Agg Inventory_Box_All Cities.png'))
plt.show()
```
